```python
import yfinance as yf
import pandas as pd

def halal_screener_agent(ticker):
    print(f"ğŸ¤– AI Agent: Analyzing stock {ticker}...")

    # 1. Fetch Company Data
    stock = yf.Ticker(ticker)
    info = stock.info

    # Attempt to fetch the Balance Sheet
    try:
        balance_sheet = stock.balance_sheet

        # Check if data exists
        if balance_sheet.empty:
            return "âš ï¸ Sorry, insufficient financial data available for this stock."

        # 2. Extract Key Figures (Latest Fiscal Year)
        # We need Total Debt and Total Assets
        total_debt = balance_sheet.loc['Total Debt'].iloc[0] if 'Total Debt' in balance_sheet.index else 0
        total_assets = balance_sheet.loc['Total Assets'].iloc[0]

        # Market Cap (Optional: Some standards use Market Cap instead of Total Assets)
        market_cap = info.get('marketCap', 0)

        # 3. Calculate Sharia Compliance Ratio (Financial Screen)
        # Rule: Interest-bearing debt should not exceed 33% of Total Assets.
        debt_ratio = (total_debt / total_assets) * 100

        # 4. Decision Making (The Logic)
        print("\n--- ğŸ“Š Financial Compliance Report ---")
        print(f"ğŸ¢ Company: {info.get('longName', ticker)}")
        print(f"âš™ï¸ Sector: {info.get('sector', 'Unknown')}")
        print(f"ğŸ’° Total Debt: ${total_debt:,.0f}")
        print(f"ğŸ›ï¸ Total Assets: ${total_assets:,.0f}")
        print(f"ğŸ“ˆ Debt-to-Assets Ratio: {debt_ratio:.2f}%")

        print("\n--- âš–ï¸ AI Agent Verdict ---")

        if debt_ratio < 33:
            print("âœ… (PASS) The stock passes the financial screening.")
            print("Reason: Debt ratio is below the 33% threshold.")
            print("Note: Please manually verify that the company's core activity is permissible (e.g., not Alcohol, Gambling, or Conventional Banking).")
        else:
            print("âŒ (FAIL) The stock does NOT pass the financial screening.")
            print(f"Reason: Debt ratio ({debt_ratio:.2f}%) exceeds the permissible limit of 33%.")

    except Exception as e:
        print(f"Error during analysis: {e}")

# --- Agent Testing ---
# Try different tickers here:
# TSLA = Tesla (Usually Passes)
# JPM = JPMorgan Chase (Conventional Bank - High Debt)
ticker_symbol = input("Enter Stock Ticker (e.g., TSLA or AAPL): ").upper()
halal_screener_agent(ticker_symbol)
```
